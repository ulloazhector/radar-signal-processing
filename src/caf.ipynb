# Cross-Ambiguity Function

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
```

```python
N = 80
F = 80
n = np.linspace(-4, 4, N)
f = np.linspace(-4, 4, F)
```

```python
def triangle(x):
    return 1 - np.abs(x)
```

```python
m, k = np.meshgrid(n, f)
```

```python
def square(n, T):
    return np.where(np.logical_and(np.abs(n) >= 0, np.abs(n) < T), 1, 0)
```

```python
def caf(xr, xe):
    values = np.zeros((N,F), dtype = 'complex_')
    for m in range(N):
        for k in range(F):
            values[m][k] = 0 + 1j*0
            for n in range(N):
                values[m][k] += xe[n] * np.conjugate(xr[n-m]) * np.exp(-1j*2*np.pi*k*n/N)
    return values


# caf = triangle(m)*triangle(k)
# fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
# surf = ax.plot_surface(m, k, caf)
```

```python
x1 =square(n, 1)
```

```python
caf_calc = caf(x1, x1)
```

```python
plt.plot(n, x1);
```

```python
plt.plot(n, np.roll(np.abs(caf_calc[:,int(N/2)]), int(N/2)));
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_dpi(200)
ax = fig.add_subplot(projection='3d')

# Grab some test data.
X, Y = np.meshgrid(n, f)
zs = np.roll(np.roll(np.array(caf_calc), int(N/2), axis=0), int(N/2), axis=1 )
Z = zs.reshape(X.shape)

# Plot a basic wireframe.
ax.plot_surface(X, Y, np.abs(Z))
plt.show()
```
